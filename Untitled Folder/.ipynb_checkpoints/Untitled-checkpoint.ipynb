```python
import numpy

def indices(list, filtr=lambda x: bool(x)):
    return [i for i,x in enumerate(list) if filtr(x)]

def trigen(t, Tw, skew):
    Ta=float(Tw)/float(2*skew)
    y = [0 for x in range(len(t))] 
    index1 = indices(t, lambda x: x>(-Tw/2) and x<Ta)
    index2 = indices(t, lambda x: x<(-Tw/2) and x>Ta)
    index3 = indices(t, lambda x: x==Ta) 
    for i in index1:
        rez = (2.0*t[i]+Tw)/(Tw*(1.0+skew));
        y.insert(i, rez)
        print(y[i])
    for i in index2:
        rez = (2.0*t[i]+Tw)/(Tw*(1.0+skew));
        y.insert(i, rez)
        print(y[i])
    for i in index3:
        y[i] = 1.0
    return y

fs = 1
t = np.arange(-10,9,1/fs)
w = 10
x = trigen(t, w, 1)# trinagular pulse function 
plt.plot(x)
print(len(x))
```

```python
import math
def my_dft_zeroPadd(s,N):
    S=len(s)
   
    if S < N:
        x = [0 for a in range(N)]
        for b in range(S):
            x[b] = s[b]
    else:
        N = S;
        x = s

    sig = [0 for x in range(N)]
    for k in range(N):
        for n in range(N):
            summ = (x[n]*np.exp((-1j*(2*np.pi*(k)*(n)))/N))
            sig[k] = sig[k]+summ
    return sig
        
def my_invdft(s):
    N = len(s)
    sig = [0 for x in range(N)]
    for k in range(N):
        for n in range(N):
            summ = (s[n]*np.exp((1j*2*np.pi*k*n)/N))
            sig[k] = sig[k]+summ
        sig[k] = float(1/N)*sig[k]
    return sig


#%% Create one sinusoid f - 100
Fs = 100;
t  = np.arange(0,0.5,1/Fs) ;
A = 1;
f1 = 30;
phi = 0;
s_c1 = A * np.sin( 2*np.pi*f1*t + phi ) ;
#%%
x = s_c1;
plt.plot(x)
lpad = 2*len(s_c1);
xdft = my_dft_zeroPadd(x,lpad);
xdft = xdft[0:int(len(xdft)/2+1)];
for i in range(len(xdft)):
    xdft[i] = xdft[i]/len(x);
xdft[1:-2] = 2*xdft[1:-2];
freq = np.arange(0,99,Fs/(2*len(x)))  
print(freq)
print(len(freq))
plt.plot(freq,xdft);
#hold on;
o = [1 for a in range(99)]
#plt.plot(freq,o)

invDftSig = my_invdft(xdft);
plt.plot(freq[:50],invDftSig[:50]);
#xlabel('Hz'); ylabel('Amplitude');
#hold off;

```

```python
def indices(list, filtr=lambda x: bool(x)):
    return [i for i,x in enumerate(list) if filtr(x)]

def trigen(t, Tw, skew):
    Ta=float(Tw)/float(2*skew)
    y = [0 for x in range(len(t))] 
    index1 = indices(t, lambda x: x>(-Tw/2) and x<Ta)
    index2 = indices(t, lambda x: x<(-Tw/2) and x>Ta)
    index3 = indices(t, lambda x: x==Ta) 
    for i in index1:
        rez = (2.0*t[i]+Tw)/(Tw*(1.0+skew));
        y.insert(i, rez)
    for i in index2:
        rez = (2.0*t[i]+Tw)/(Tw*(1.0+skew));
        y.insert(i, rez)
    for i in index3:
        y[i] = 1.0
    return y
```

```python
fs = 1
t = np.arange(-10,9,1/fs)
w = 10
x = trigen(t, w, 1)# trinagular pulse function 
plt.plot(x)
plt.xlabel('Time [samples]')
plt.ylabel('Amplitude')
```

```python
Fs = 100 ; ti = 0 ; tf = 0.5 ;
t  = np.arange(ti,tf,1/Fs)
f = 5 ;
A = 10 ; 
phi = 0 ;

# simulation of a sinusoidal signal s(t) at continuous time : s_c
s_c = A * np.sin( 2*np.pi*f*t + phi ) ;
phi = np.pi/2;
s_c2 = A * np.sin( 2*np.pi*f*t + phi ) ;
phi = np.pi
s_c3 = A * np.sin( 2*np.pi*f*t + phi ) ;
plt.plot(t,s_c)
plt.plot(t,s_c2)
plt.plot(t,s_c3)
plt.xlabel('continuous time');
plt.ylabel('Amplitude');
plt.title('continuous sinusoidal signal with different phase');
```

```python
n  = np.arange(0,(len(t)-1),1) ;
# sinusoidal signal sinusoidal s(t) at discrete time : s_d
s_d = A * np.sin( 2*np.pi*f/Fs*n + phi ) ;

plt.plot(s_d)
plt.stem(s_d);
plt.xlabel('discrete time');
plt.ylabel('Amplitude');
plt.title('discrete sinusoidal signal');
```

```python
#%% Create one sinusoid f - 100
Fs = 256;
t  = np.arange(0,0.5,1/Fs) ;
A = 1;
f1 = 100;
phi = 0;
s_c1 = A * np.sin( 2*np.pi*f1*t + phi ) ;
#%% create another sinusoid f - 156
f2 = 156;
s_c2 = A * np.sin( 2*np.pi*f2*t + phi ) ;
#%% Sum both signals
sig = s_c1+s_c2;

plt.xlabel('continuous time');
plt.ylabel('Amplitude');
plt.title('Phase opposite signals s1 and s2, can be retrieved by the relation sin(a-b)');
plt.plot(t,s_c1,t,s_c2)
```

```python
def tic():
    #Homemade version of matlab tic and toc functions
    import time
    global startTime_for_tictoc
    startTime_for_tictoc = time.time()

def toc():
    import time
    if 'startTime_for_tictoc' in globals():
        print ("Elapsed time is " + str(time.time() - startTime_for_tictoc) + " seconds.")
    else:
        print ("Toc: start time not set")
```

```python
def my_dft_zeroPadd(s,N):
    S=len(s)
   
    if S < N:
        x = [0 for a in range(N)]
        for b in range(S):
            x[b] = s[b]
    else:
        N = S;
        x = s

    sig = [0 for x in range(N)]
    for k in range(N):
        for n in range(N):
            summ = (x[n]*np.exp((-1j*(2*np.pi*(k)*(n)))/N))
            sig[k] = sig[k]+summ
    return sig
        
def my_invdft(s):
    N = len(s)
    sig = [0 for x in range(N)]
    for k in range(N):
        for n in range(N):
            summ = (s[n]*np.exp((1j*2*np.pi*k*n)/N))
            sig[k] = sig[k]+summ
        sig[k] = float(1/N)*sig[k]
    return sig
```

```python
#%% Create one sinusoid f - 100
Fs = 100;
t  = np.arange(0,0.5,1/Fs) ;
A = 1;
f1 = 30;
phi = 0;
s_c1 = A * np.sin( 2*np.pi*f1*t + phi ) ;
x = s_c1;
plt.plot(x)
plt.xlabel('Time');
plt.ylabel('Amplitude');
```

```python
lpad = 2*len(s_c1);
tic()
xdft = my_dft_zeroPadd(x,lpad);
toc()
plt.plot(np.real(xdft))
plt.xlabel('frequency');
plt.ylabel('Amplitude');
plt.title('Discrete Fourier Transformation')
```

```python
#InvDFT
freq = np.arange(0,50,Fs/(2*len(x)))
tic()
invDftSig = my_invdft(xdft);
toc()
plt.plot(freq,invDftSig[:50]);
plt.xlabel('Time');
plt.ylabel('Amplitude');
plt.title('Inverse Discrete Fourier Transformation')
```

```python
tic()
pythFFT = np.fft.fft(x)
toc()
plt.plot(np.real(pythFFT))
plt.xlabel('frequency');
plt.ylabel('Amplitude');
plt.title('Discrete Fourier Transformation')
```

```python
tic()
pythIFFT = np.fft.ifft(pythFFT)
toc()
plt.plot(np.real(pythIFFT))
plt.xlabel('Time');
plt.ylabel('Amplitude');
plt.title('Discrete Fourier Transformation')
```

```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
phase = -10
lenght = 30 
amplitude = 10 
x = np.arange(0,100,0.1)
```

```python
def triang(x,phase,lenght, amplitude):
    alpha = (amplitude)/(lenght/2)
    return -amplitude/2+amplitude*((x-phase)%lenght==lenght/2) \
+alpha*((x-phase)%(lenght/2))*((x-phase)%lenght<lenght/2) \
+(amplitude-alpha*((x-phase)%(lenght/2)))*((x-phase)%lenght>lenght/2)
```

```python
tr = triang(x,phase,lenght, amplitude)
plt.plot(tr)
```
